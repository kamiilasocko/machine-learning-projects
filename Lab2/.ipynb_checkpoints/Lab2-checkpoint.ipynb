W danych widzę 14 kolumn, w których opisane są konkretne osoby, biorące udział w rejsie Titanica. Mamy podane m.in wiek osób, imię, płeć, numery ich biletów czy przeżyły czy nie.. W niektórych miejscach mamy wartości "?" - zapewne są to braki danych. 

**zadanie 3**

```python
pd.isnull(df).sum() #mamy 0 wartości w tym momencie
df=df.replace("?",np.nan)
pd.isnull(df).sum()
```

```python
pd.isnull(df).mean()
```

Na początku należało zmienić znak "?" występujący w danych na wartości NaN, aby można było zliczyć prawidłowo. Przed zmianą nie liczyło prawidłowo, ponieważ dostawałam wartości 0, czyli tak jakby nie było wogóle braków danych. Jeśli mowa o % brakujących danych w zależności od danej zmiennej to największy procent występuje dla zmiennej body(ok. 91%), boat(63%), cabin(77%), home.dest(43%) i age(20%). Dla zmiennych - fare i embarked występują bardzo małe braki w danych, zatem można je usunąć. W reszcie zmiennych braki nie występują

**zadanie 4**

```python
nan_values=df.isna().sum().sum()
dim=df.size
percent_ofNaN=round(nan_values/dim*100,2)
print(percent_ofNaN)
```

W całym zbiorze jest 21% wartości NaN zatem jest to większa liczba niż 5%, więc braki nie są całkowicie przypadkowe - MCAR.
Moim zdaniem są to braki nieprzypadkowe - MNAR, ponieważ kolumny, w których występują braki danych są często związane z tym czy ktoś przeżył czy też nie. 

```python
cols=df.columns[df.isna().any()]
df[cols].isna().groupby(df["survived"]).mean()
```

```python
df['ageNull']=np.where(df['age'].isnull(),1,0)
df.groupby(['survived'])['ageNull'].mean()
```

Interpretacja: braki danych w kolumnie age dla osób, które nie przeżyły wynoszą około 23%, a dla osób które przeżyły około 15%. Analogiczna interpretacja dla reszty wyników.

Sprawdziłam zatem czy nie ma osób o tym samym imieniu - okazało się że pojawiły się dwukrotnie takie przypadki. 
Dla etykiety ticket również jest duża ilość unikalnych wartości - być może kilka osób podróżowało na jednym bilecie. 

```python
#przyjmuje że powyżej 100 unikalnych wartości to zmienne o dużej mocy-kardynalności 
high={}
low={}
for i in df.columns:
    car=len(df[i].unique())
    if(car>100):
        high[i]=car
    else:
        low[i]=car
print('high: '+str(high))
print('low: '+str(low))
```
